Sử dụng hàm numpy.array (object, dtype=None, *, copy=True, order='K', subok=False, ndmin=0, like=None)

- object : Một mảng bất kỳ đối tượng nào hiển thị giao diện mảng, một đối tượng có phương thức array trả về một mảng hoặc bất kỳ chuỗi (lồng nhau) nào. Nếu đối tượng là một vô hướng, một mảng 0 chiều chứa đối tượng được trả về.
- kiểu dữ liệu (dtype) : tùy chọn. Kiểu dữ liệu mong muốn cho mảng. Nếu không được đưa ra, thì loại sẽ được xác định là loại tối thiểu cần thiết để giữ các đối tượng trong chuỗi.
- copy : bool, tùy chọn. Nếu đúng (mặc định), thì đối tượng được sao chép. Nếu không, một bản sao sẽ chỉ được tạo ra nếu array trả về một bản sao, nếu obj là một chuỗi lồng nhau hoặc nếu một bản sao là cần thiết để đáp ứng bất kỳ yêu cầu nào khác.
- subok bool : tùy chọn. Nếu True, thì các lớp con sẽ được chuyển qua, nếu không mảng được trả về sẽ bị buộc phải là mảng lớp cơ sở (mặc định).
- int ndmin : tùy chọn. Chỉ định số thứ nguyên tối thiểu mà mảng kết quả phải có. Những chiếc sẽ được chờ trước về hình dạng khi cần thiết để đáp ứng yêu cầu này.
- like : array_like. Đối tượng tham chiếu để cho phép tạo mảng không phải là mảng NumPy. Nếu một mảng giống như được truyền vào likehỗ trợ array_functiongiao thức, kết quả sẽ được xác định bởi nó. Trong trường hợp này, nó đảm bảo việc tạo một đối tượng mảng tương thích với đối tượng được truyền vào thông qua đối số này.

```python
import numpy as np
```

```python
A = np.array([[5],[7],[14]])
print("khoi tao diem A trong khong gian 3 chieu: ")
print(A)
```
