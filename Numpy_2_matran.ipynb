Sử dụng hàm random.randint(low, high=None, size=None, dtype=int)

- low : Các số nguyên thấp nhất sẽ được rút ra từ phân phối (trừ khi high=None, trong trường hợp đó, tham số này nằm trên một số nguyên cao nhất như vậy).
- high : Nếu được cung cấp, một bên trên số nguyên lớn nhất sẽ được rút ra từ phân phối
- size : hình dạng đầu ra

```python
A = np.random.randint(1,high = 10,size=([7,7]))
print('Ma tran vuong A cap 7x7: \n')
print(A)
```

Tạo ma trận chéo : Sử dụng hàm numpy.fill_diagonal(a, val, wrap=False)

- a : mảng
- val : giá trị cần ghi trên đường chéo

```python
A = np.zeros([6,6])
np.fill_diagonal(A,val=15)
print("ma tran cheo A: ")
print(A)
```
