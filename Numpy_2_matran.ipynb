S dng hm random.randint(low, high=None, size=None, dtype=int)

- low : Cc s nguyn thp nht s c rt ra t phn phi (tr khi high=None, trong trng hp , tham s ny nm trn mt s nguyn cao nht nh vy).
- high : Nu c cung cp, mt bn trn s nguyn ln nht s c rt ra t phn phi
- size : hnh dng u ra

```python
A = np.random.randint(1,high = 10,size=([7,7]))
print('Ma tran vuong A cap 7x7: \n')
print(A)
```

To ma trn cho : S dng hm numpy.fill_diagonal(a, val, wrap=False)

- a : mng
- val : gi tr cn ghi trn ng cho

```python
A = np.zeros([6,6])
np.fill_diagonal(A,val=15)
print("ma tran cheo A: ")
print(A)
```
